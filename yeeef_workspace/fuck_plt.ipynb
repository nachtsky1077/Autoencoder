```python
from matplotlib import pyplot as plt
import numpy as np
```

first, let's generate some fake data for illustration

```python
cors = np.random.rand(500, 2)
```

```python
plt.scatter(cors[:, 0], cors[:, 1], c=[0.1, 0.2, 0.3])
plt.show()
```

```python
plt.scatter(cors[:, 0], cors[:, 1], c=[[0.1, 0.2, 0.3]])
plt.show()
```

```python
plt.scatter(cors[:, 0], cors[:, 1], c=[[1, 2, 3]])
plt.show()
```

上面这个错误，乍一看还以为是 size 不合适，那么实际上呢

```python
plt.scatter(cors[:, 0], cors[:, 1], c=np.array([1, 2, 3] * 500).reshape(-1, 3))
plt.show()
```

看吧，如此奇葩的报错，谁能顶得住？

所以，在 plt 中，当用 rgb, rbga 设置颜色时，都要用 [0, 1] 区间内的 value, 通过 matplotlib.cm 可以方便地为我们产生这个需求, 需要注意的是仍需要用 [0, 1] 之间的 value 作为参数

```python
import matplotlib.cm as cm
```

```python
colors = cm.rainbow(np.linspace(0, 1, 10))
```

```python
colors
```

如果用 0～2 虽然不会报错，但是后几个颜色是完全相同的

```python
colors = cm.rainbow(np.linspace(0, 2, 10))
colors
```
